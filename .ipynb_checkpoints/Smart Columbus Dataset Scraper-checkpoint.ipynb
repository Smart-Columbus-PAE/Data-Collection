```python
def webscrape(website):

    from selenium import webdriver
    import time
    from selenium.webdriver.chrome.options import Options
    import re


    options = Options()
    options.headless = True

    EXE_PATH = r'/Applications/chromedriver'
    driver = webdriver.Chrome(executable_path=EXE_PATH, options=options)
    driver.set_window_size(1440, 900)
    driver.get(website)
    
    last_modified= []
    file_type = []
    description = []
    url_final = []
    title_final =[]

    #for a in driver.find_elements_by_xpath(".//div[@class='result-count']"):
    #    print(a)
    #    num = str(a.text)
    
    #num = re.sub("[^0-9]", "", num)
    #num2 = round(int(num)/10,0)+1
        
    j=0
    
    if 'apiAccessible=true' in website:
        num=72
    else:
        num=320
    
    for x in range(1,num+1):

        site = website+str(x)
        url_temp = []
        title_temp = []
        
        driver.get(site)
        time.sleep(5)
        z=0
        for a in driver.find_elements_by_class_name('title'):
            z+=1
            if z>1:
                url_final.append(str(a.get_attribute('href')))
                title_final.append(str(a.text))

        for a in driver.find_elements_by_xpath(".//div[@class='last-modified']"):
            last_modified.append(str(a.text))

        for a in driver.find_elements_by_xpath(".//span[@class='file-type']"):
            file_type.append(str(a.text))

        for a in driver.find_elements_by_xpath(".//div[@class='description']"):
            description.append(str(a.text))
        
        time.sleep(2)
        j+=1
        if j%20==0:
            print(str(j)," pages scraped, ", str(len(url_final)), " datasets collected")
            print(url_final[-1])

    driver.close()
    return last_modified, file_type , description , url_final ,title_final

```

```python
def webscrape_onesite(website):

    from selenium import webdriver
    import time
    from selenium.webdriver.chrome.options import Options
    import re


    options = Options()
    options.headless = True

    EXE_PATH = r'/Applications/chromedriver'
    driver = webdriver.Chrome(executable_path=EXE_PATH, options=options)
    driver.set_window_size(1440, 900)
    driver.get(website)
    
    last_modified= []
    file_type = []
    description = []
    url_final = []
    title_final =[]
    

    site = website
    url_temp = []
    title_temp = []

    driver.get(site)
    time.sleep(5)
    z=0
    for a in driver.find_elements_by_class_name('title'):
        z+=1
        if z>1:
            url_final.append(str(a.get_attribute('href')))
            title_final.append(str(a.text))

    for a in driver.find_elements_by_xpath(".//div[@class='last-modified']"):
        last_modified.append(str(a.text))

    for a in driver.find_elements_by_xpath(".//span[@class='file-type']"):
        file_type.append(str(a.text))

    for a in driver.find_elements_by_xpath(".//div[@class='description']"):
        description.append(str(a.text))

    driver.close()
    return last_modified, file_type , description , url_final ,title_final

```

```python
for x in ['https://discovery.smartcolumbusos.com/?q=&sort=name_asc&apiAccessible=true&page=72',
         'https://discovery.smartcolumbusos.com/?q=&sort=name_asc&apiAccessible=false&page=320']:

    last_modified, file_type , description , url_final ,title_final = webscrape_onesite(x)

    df_temp = pd.DataFrame({
        'Title':title_final,
        'File Type':file_type,
        'Last Modified Date':last_modified,
        'Description':description,
        'URL':url_final
    }) 
    
    agency = []
    new_dates = []

    for x in df_temp['URL']:
        y = x[46:]
        z=y[:y.find("/")]
        agency.append(z)

    for x in df_temp['Last Modified Date']:
        y = x[:12]
        if y == "Date not pro":
            y=""
        if y == "Updates to r":
            y=""
        new_dates.append(y)
    
    df_temp['Agency']=agency    
    df_temp['Last Modified Date New']=new_dates
    df_temp['date pulled']=date.today()
    df=df.append(df_temp)
    print("done")
```

```python
def deep_scrape(website):
    from selenium import webdriver
    import time
    from selenium.webdriver.chrome.options import Options
    import re


    options = Options()
    options.headless = True

    EXE_PATH = r'/Applications/chromedriver'
    driver = webdriver.Chrome(executable_path=EXE_PATH, options=options)
    driver.set_window_size(1440, 900)
    driver.get(website)

    time.sleep(5)
    z=0
    temp_list = []
    api_query_num=''
    download_num=''
    
    for a in driver.find_elements_by_xpath(".//div[@class='rt-td']"):
        temp_list.append(a.text)
        if z==1:
            api_query_num = a.text
            z=0
        if z==2:
            download_num = a.text
            z=0
        if a.text == "Count - API Queries":
            z=1
        if a.text == "Count - Downloads":
            z=2
    
    return temp_list, api_query_num, download_num
```
